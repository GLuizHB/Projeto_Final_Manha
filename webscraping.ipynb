## Manipulação de strings e Intro (ex. Regular)

- Espreções Regulares auxiliam nas validações

```python
df['times'] = df['times'].str.replace('>>','',regex=False).str.strip()
df[['times','V']].head(10)
```

```python
import matplotlib as plt
```

```python
grafico_percentual_vitorias = df.groupby('times')[['percentual']].sum().sort_values('percentual')
```

```python
grafico_percentual_vitorias.plot(kind='barh',figsize=(14,10),color='purple')
```

### Grafico Pizza, para um time expecifico

```python
# Suponha que 'df' seja o seu DataFrame e que você queira filtrar um time específico.
time_especifico = 'Bahia'

# Filtrando o DataFrame para o time específico
dados_time = df[df['times'] == 'Bahia']

# Somando as vitórias, derrotas e empates para o time específico
soma_dados_time = dados_time[['D', 'V', 'E']].sum()

# Criando o gráfico de pizza
soma_dados_time.plot(kind='pie', figsize=(14, 10), autopct='%1.1f%%', colors=['red', 'green', 'grey'])

plt.title(f'Distribuição de Resultados do Time {time_especifico}')
plt.ylabel('')  # Remove o rótulo do eixo y
plt.show()
```
